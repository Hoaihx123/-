```python
import pandas as pd
table = pd.read_csv("data2.csv", sep=',')
df_1 = pd.DataFrame(table)
df_1
```

```python
from sklearn import preprocessing
scaler_std = preprocessing.StandardScaler()
x = scaler_std.fit_transform(df_1[['USD']])
df_1['USD'] = x[0:]
x = scaler_std.fit_transform(df_1[['CAN']])
df_1['CAN'] = x[0:]
x = scaler_std.fit_transform(df_1[['EUR']])
df_1['EUR'] = x[0:]
df_1
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
probplot = sm.ProbPlot(df_1['USD'])
probplot.qqplot(line='r')
probplot.ppplot(line='45')
plt.show
```

```python
fix, ax = plt.subplots()
ax.vlines(df_1['USD'].mean(), 0, df_1['USD'].size, color='red')
ax.vlines(df_1['USD'].median(), 0, df_1['USD'].size, color='green')
ax.vlines(df_1['USD'].mode()[0], 0, df_1['USD'].size, color='blue')
df_1['USD'].plot.hist()
plt.show()
```

```python
fix, ax = plt.subplots()
ax.vlines(df_1['CAN'].mean(), 0, df_1['CAN'].size, color='red')
ax.vlines(df_1['CAN'].median(), 0, df_1['CAN'].size, color='green')
ax.vlines(df_1['CAN'].mode()[0], 0, df_1['CAN'].size, color='blue')
df_1['CAN'].plot.hist()
plt.show()
```

```python
fix, ax = plt.subplots()
ax.vlines(df_1['EUR'].mean(), 0, df_1['EUR'].size, color='red')
ax.vlines(df_1['EUR'].median(), 0, df_1['EUR'].size, color='green')
ax.vlines(df_1['EUR'].mode()[0], 0, df_1['EUR'].size, color='blue')
df_1['EUR'].plot.hist()
plt.show()
```
