```python
import pandas as pd
table = pd.read_csv(r"/content/winequality-red.dat", comment='@', sep=',', header=None)
df = pd.DataFrame(table)
df.columns = [
    "fixed_acidity", "volatile_acidity", "citric_acid",
    "residual_sugar", "chlorides", "free_sulfur_dioxide",
    "total_sulfur_dioxide", "density", "pH", "sulphates",
    "alcohol", "quality"
]
df.head(100)
```

```python
df['quality'].value_counts()
```

```python
from sklearn import preprocessing
scaler_std = preprocessing.StandardScaler()
x = scaler_std.fit_transform(df[["fixed_acidity", "volatile_acidity", "citric_acid",
    "residual_sugar", "chlorides", "free_sulfur_dioxide",
    "total_sulfur_dioxide", "density", "pH", "sulphates",
    "alcohol"]])
df[["fixed_acidity", "volatile_acidity", "citric_acid",
    "residual_sugar", "chlorides", "free_sulfur_dioxide",
    "total_sulfur_dioxide", "density", "pH", "sulphates",
    "alcohol"]] = x
df
```

```python
from sklearn.model_selection import train_test_split
train_data, temp_data = train_test_split(df, test_size=0.5, random_state=2406)
test_data, val_data = train_test_split(temp_data, test_size=0.4, random_state=2406)
train_data = train_data.reset_index()
test_data = test_data.reset_index()
val_data = val_data.reset_index()
print(train_data.shape, test_data.shape, val_data.shape)
```

```python
test_data
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
```

```python
def tree_train(x_train, y_train, x_test, y_test):
  results = []
  alphas = [0.005, 0.01, 0.015, 0.02, 0.025, 0.03, 0.035, 0.2, 0.8]
  criteria = ['gini', 'entropy']
  for criterion in criteria:
      for alpha in alphas:
          clf = DecisionTreeClassifier(criterion=criterion, ccp_alpha=alpha, random_state=2406)
          clf.fit(x_train, y_train)
          y_pred = clf.predict(x_test)
          y_proba = clf.predict_proba(x_test)
          accuracy = accuracy_score(y_test, y_pred)
          precision = precision_score(y_test, y_pred, average='weighted')
          recall = recall_score(y_test, y_pred, average='weighted')
          roc_auc = roc_auc_score(y_test, y_proba, multi_class='ovr', average='macro')

          results.append({
              'criterion': criterion,
              'alpha': alpha,
              'accuracy': accuracy,
              'precision': precision,
              'recall': recall,
              'roc_auc': roc_auc
          })
  return pd.DataFrame(results)
```

```python
x_train = train_data[["fixed_acidity", "volatile_acidity", "citric_acid",
      "residual_sugar", "chlorides", "free_sulfur_dioxide",
      "total_sulfur_dioxide", "density", "pH", "sulphates",
      "alcohol"]]
y_train = train_data["quality"]
x_test = test_data[["fixed_acidity", "volatile_acidity", "citric_acid",
      "residual_sugar", "chlorides", "free_sulfur_dioxide",
      "total_sulfur_dioxide", "density", "pH", "sulphates",
      "alcohol"]]
y_test = test_data['quality']
res = tree_train(x_train, y_train, x_test, y_test)
```

```python
from sklearn import svm
gammas = ['scale', 'auto']
coef0s = [0, 1, 2]
degrees = [2, 3, 4]
Cs = [0.1, 1, 10]
```

```python
def svm_train(x_train, y_train, x_test, y_test):
  results = []
  for krl in ['linear', 'poly', 'rbf', 'sigmoid']:
    for gamma in gammas:
      for coef0 in coef0s:
        for degree in degrees:
          for C in Cs:
            clf = svm.SVC(kernel = krl, gamma = gamma, coef0 = coef0, degree = degree, C = C, probability=True)
            clf.fit(x_train, y_train)
            y_pred = clf.predict(x_test)
            y_proba = clf.predict_proba(x_test)
            accuracy = accuracy_score(y_test, y_pred)
            precision = precision_score(y_test, y_pred, average='weighted')
            recall = recall_score(y_test, y_pred, average='weighted')
            roc_auc = roc_auc_score(y_test, y_proba, multi_class='ovr', average='macro')
            results.append({
                'krl': krl,
                'gammas': gamma,
                'coef0s': coef0,
                'degrees': degree,
                'Cs': C,
                'accuracy': accuracy,
                'precision': precision,
                'recall': recall,
                'roc_auc': roc_auc
          })
  return pd.DataFrame(results)
```

```python
res = svm_train(x_train, y_train, x_test, y_test)
```
