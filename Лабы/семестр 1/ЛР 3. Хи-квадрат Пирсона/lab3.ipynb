```python
import matplotlib.pyplot as plt
fix, ax = plt.subplots()
ax.vlines(df['Income'].mean(), 0, df['Income'].size, color='red')
ax.vlines(df['Income'].median(), 0, df['Income'].size, color='green')
ax.vlines(df['Income'].mode()[0], 0, df['Income'].size, color='blue')
df['Income'].plot.hist()
plt.show()
```

```python
df['Income']=df[['Income']].applymap(lambda x: np.log(x+1))
df
```

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
model = ols('Income ~ C(EmploymentField) + C(EmploymentStatus) + C(Gender)+ C(JobPref)+ C(JobWherePref)+ C(MaritalStatus)', data=df).fit()
anova_lm(model)
```

```python
from statsmodels.stats.stattools import durbin_watson
dw_stat = durbin_watson(model.resid)
dw_stat
```

```python
from statsmodels.stats.diagnostic import het_breuschpagan
het_breuschpagan(model.resid, model.model.exog)
```
